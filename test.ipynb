```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('./images//dl2_city_conceptart.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
gray_img = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
gray_img = cv.applyColorMap(gray_img, cv.COLORMAP_JET)
plt.imshow(gray_img)
plt.show()
```
