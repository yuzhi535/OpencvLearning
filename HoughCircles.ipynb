```python
# import package
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
file_name = 'target.jpg'
img = cv.imread(file_name)

if img is None:
    print(os.listdir())
    print('failure')
else:
    plt_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.imshow(plt_img)
    plt.show()
```

```python
gray_img = cv.imread('target.jpg', 0)
gray_img = cv.cvtColor(gray_img, cv.COLOR_GRAY2BGR)

circles = cv.HoughCircles(gray_img, cv.HOUGH_GRADIENT, 1, 20,
                          param1=50, param2=30, minRadius=0, maxRadius=0)

circles = np.uint16(np.around(circles))
for i in circles[0, :]:
    # draw the outer circle
    cv.circle(img, (i[0], i[1]), i[2], (0, 255, 0), 2)
    # draw the center of the circle
    cv.circle(img, (i[0], i[1]), 2, (0, 0, 255), 3)

plt_img = cv.cvtColor(gray_img, cv.COLOR_GRAY2BGR)
plt_img = cv.cvtColor(plt_img, cv.COLOR_BGR2RGB)
plt.imshow(plt_img)
plt.show()
```
