# 毛玻璃效果

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

## 显示图片

```python
path = './images/716428.png'
img = cv.imread(path)
out_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(out_img)
plt.show()
```

```python
out_img = cv.GaussianBlur(img, (0,0), 20)

out_img = cv.cvtColor(out_img, cv.COLOR_BGR2RGB)
plt.imshow(out_img)
plt.show()
```
