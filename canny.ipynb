# Canny边缘检测

## 原理
1. 去噪
2. 计算梯度，如sobel算子
3. 非极大值抑制，得到极大值，尽可能接近轮廓
4. 认为设置两个最大最小阈值，进一步去掉多余轮廓和小的噪声点
---

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('./images/target.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.title('原图')
plt.imshow(img)
plt.show()
```

```python
output = cv.Canny(img, 100, 200)
plt.imshow(output, cmap='gray')
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.title('origin')
plt.imshow(img)
plt.subplot(1, 2, 2)
plt.title('result')
plt.imshow(output, cmap='gray')
plt.show()
```

```python
fun_img = cv.bitwise_not(output)
plt.imshow(fun_img, cmap='gray')
plt.show()
```
