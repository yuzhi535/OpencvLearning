```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('./images/target.jpg', 0)
target_img = cv.Sobel(img, cv.CV_64F, 1, 0, ksize=5)
target_img = np.absolute(target_img)
target_img = np.uint8(target_img)
plt.title('origin')
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('result')
plt.subplot(1, 2, 2)
plt.imshow(target_img, cmap='gray')
plt.show()
```
